```python
import os
import matplotlib.pyplot as plt

labels_dir = '/content/yolov5/runs/detect/video_results2/labels'
frame_counts = []
crowd_levels = []

for i in range(count):
    label_file = f"{labels_dir}/frame_{i:04d}.txt"
    if os.path.exists(label_file):
        with open(label_file, "r") as f:
            persons = len(f.readlines())
    else:
        persons = 0

    frame_counts.append(persons)

    # Classify crowd density
    if persons <= 10:
        crowd = "Low"
    elif persons <= 30:
        crowd = "Medium"
    else:
        crowd = "High"
    crowd_levels.append(crowd)

print("âœ… People counted and crowd levels classified.")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(range(len(frame_counts)), frame_counts, marker='o', label='People Count', color='dodgerblue')
plt.title("People Count per Frame")
plt.xlabel("Frame Number")
plt.ylabel("Number of People")
plt.grid(True)
plt.legend()
plt.show()
```

```python
density_map = {"Low": 1, "Medium": 2, "High": 3}
density_values = [density_map[cl] for cl in crowd_levels]

plt.figure(figsize=(12, 5))
plt.plot(range(len(density_values)), density_values, marker='s', color='green', label='Crowd Density')
plt.yticks([1, 2, 3], ['Low', 'Medium', 'High'])
plt.title("Crowd Density per Frame")
plt.xlabel("Frame Number")
plt.ylabel("Crowd Density")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd

csv_data = {
    "Frame": [f"frame_{i:04d}.jpg" for i in range(len(frame_counts))],
    "People Count": frame_counts,
    "Crowd Density": crowd_levels
}

df = pd.DataFrame(csv_data)

csv_path = "/content/crowd_analysis_results.csv"
df.to_csv(csv_path, index=False)

from google.colab import files
files.download(csv_path)
```
